```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
#Note: Also add title, xlabel, and ylabel to the plot.

import matplotlib.pyplot as plt
plt.scatter(x,y,c=x)
plt.xlabel("X data")
plt.ylabel("Y data")
plt.title("Scatter plot")
```

```python
#Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title('Plot 1')
plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title('Plot 2')
plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title('Plot 3')
plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title('Plot 4')
```

```python
#Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
# Bar Plot:
#Ans: A bar plot is a graphical representation of categorical data with rectangular bars. Each bar's 
#length or height is proportional to the values they represent. Bar plots are commonly used to display and compare the frequency, count, or other 
#aggregate measures of different categories. They are suitable for showing the distribution or relationship between categorical variables.

#Why is it used:
#Bar plots are useful for visualizing and comparing the magnitude of different categories. They provide a straightforward way to convey 
#information about the distribution of data and are effective in highlighting variations or trends

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.show()
```

```python
#Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
# Box Plot:
'''A box plot, also known as a box-and-whisker plot, is a statistical visualization that displays the distribution of a dataset. 
It provides a summary of the data's central tendency, spread, and potential outliers. The key components of a box plot include the 
median (central line inside the box), the interquartile range (IQR, represented by the box), and the whiskers that extend to the minimum and 
maximum values within a certain range.

Why is it used:
Box plots are used to understand the distribution of a dataset and to identify the presence of outliers. They are particularly useful for 
comparing the distribution of multiple datasets or for visualizing the spread of data in a single dataset. Box plots are effective in 
showing the skewness, central tendency, and variability of the data in a compact and informative manner.'''
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.subplot(2,2,1)
plt.boxplot(box1)
plt.subplot(2,2,2)
plt.boxplot(box2)
```
